## Brake control

### Variable input:

$$
\begin{aligned}
m - vehicle mass \\[6pt]
v - velocity m - vehicle mass \\[6pt]
t - ramp time \\[6pt]
dt - cycle time \\[6pt]
tao - time constant PT1 \\[6pt]
Kp - proportional control factor \\[6pt]
Ki - integral control factor \\[6pt]
Kd - differential control factor \\[6pt]
\end{aligned}
$$

### 1. Basis Formeln
$$
\begin{aligned}
M_{\mathrm{target}} &= F_{B\mathrm{}}\cdot r \\[6pt]
a_{target} &= \frac{v_0^2}{2S_{\text{stop}}} \\[6pt]
\end{aligned}
$$

### 2. Newton umformen:
$$
\begin{aligned}
F &= m\cdot a_{\mathrm{}} \\[6pt]
F_{B\mathrm{}} &= m\cdot a_{\mathrm{B}} - F_{W} \\[6pt]
einsetzen: 
F_{B} &= \frac{M_{B}}{r_{\text{}}} \\[6pt]
\frac{M_{B}}{r_{\text{}}} &= m\cdot a_{\mathrm{B}} - F_{W} \\[6pt]
{M_{B}} &= r\cdot(m\cdot a_{\mathrm{B}} - F_{W}) \\[6pt]
\end{aligned}
$$

### 3. Zeitabhängiges Bremsmoment (für Rampe):

$$
\begin{align*}
& M_{\mathrm{0}} - Anfangsmoment & \\[6pt]
& t_{\mathrm{i}} - Zeitinkrement & \\[6pt]
& t_{\text{R}} - Rampenzeit & \\[6pt]
\end{align*}
$$

$$
\begin{aligned}
M_{B\mathrm{}(t)} &= M_{\mathrm{0}} + \frac{M_{B}-M_{0}}{t_{\text{R}}} \cdot t_{\mathrm{i}} \\[6pt]
\end{aligned}
$$



### 4. Umrechnung Druck:

$$
\begin{aligned}
M_{B\mathrm{}} &= p_{\mathrm{}} \cdot K_{\mathrm{}} \\[6pt]
p_{B\mathrm{}} &= \frac{M_{B}}{K_{\mathrm{}}} \\[6pt]
\end{aligned}
$$


### Bedingung:

$0 \leq t \leq t_{\text{ramp}}$ and
$t > t_{\text{ramp}}$
