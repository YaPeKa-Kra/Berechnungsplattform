### Binary pack (struct)

| Code | Typ     | Bytes | Bereich      |
| ---- | ------- | ----- | ------------ |
| `B`  | uint8   | 1     | 0–255        |
| `b`  | int8    | 1     | -128–127     |
| `H`  | uint16  | 2     | 0–65535      |
| `h`  | int16   | 2     | -32768–32767 |
| `I`  | uint32  | 4     | 0–4.29e9     |
| `i`  | int32   | 4     | ±2.1e9       |
| `f`  | float   | 4     | IEEE754      |
| `d`  | double  | 8     | IEEE754      |
| `8s` | 8 Bytes | 8     | Rohdaten     |


<  = Little Endian
">  = Big Endian


our example: struct.pack("<IBB8sI", can_id, dlc, flags, data, timestamp)

therefore means:
| Reihenfolge | Code | Bedeutung     | Byte |
| ----------- | ---- | ------------- | ---- |
| 1           | `<`  | Little Endian | –    |
| 2           | `I`  | CAN-ID        | 4    |
| 3           | `B`  | DLC           | 1    |
| 4           | `B`  | Flags         | 1    |
| 5           | `8s` | Daten         | 8    |
| 6           | `I`  | Timestamp     | 4    |
